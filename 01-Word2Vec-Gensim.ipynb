# Word2Vec using Gensim
## This notebook outlines the concepts of applying Word2Vec using Gensim library

```python
import nltk
import gensim
from nltk.corpus import abc
```

```python
model = gensim.models.Word2Vec(abc.sents())
```

```python
X = list(model.wv.vocab)
X
```

```python
data = model.most_similar('science')
data
```

```python
data = model.most_similar('Australia')
data
```

```python
data = model.most_similar('market')
data
```

```python
data = model.most_similar('Coalition')
data
```
