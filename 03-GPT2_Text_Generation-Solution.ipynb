### Import the necessary libraries

```python
import torch
from pytorch_transformers import GPT2Tokenizer, GPT2LMHeadModel
```

### Loading pre-trained model tokenizer

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

### Encode text inputs

```python
text = "What is the fastest car in the"
indexed_tokens = tokenizer.encode(text)
indexed_tokens
```

### Convert indexed tokens in a Pytorch tensor

```python
tokens_tensor = torch.tensor([indexed_tokens])
tokens_tensor
```

### Load pre-trained model (weights)

```python
model = GPT2LMHeadModel.from_pretrained('gpt2')
model
```

### Set the model in evaluation mode to deactivate the Dropout modules

```python
model.eval()
```

### GPU usage

```python
# If you have a GPU, put everything on cuda by uncommenting and run this
# tokens_tensor = tokens_tensor.to('cuda')
# model.to('cuda')
```

### Predict tokens

```python
with torch.no_grad():
    outputs = model(tokens_tensor)
    predictions = outputs[0]
```

```python
outputs
```

### Get the predicted next sub-word

```python
predicted_index = torch.argmax(predictions[0, -1, :]).item()
predicted_text = tokenizer.decode(indexed_tokens + [predicted_index])
```

```python
print(predicted_text)
```

# Transformers library

## Pipeline

### Import the library

```python
from transformers import pipeline, set_seed
```

### Create a pipeline for text generation with the chosen model

```python
generator = pipeline('text-generation', model='gpt2')
```

### Set a seed to recreate / reproduce the stochastic output

```python
set_seed(42)
```

### Generate text

```python
generator("Hello, I'm a language model,", max_length=30, num_return_sequences=5)
```
